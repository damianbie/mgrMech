```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams.update({
    'axes.facecolor': 'white',
    'axes.edgecolor': 'black',
    'axes.grid': True,
    'grid.color': 'gray',
    'grid.linestyle': '-',
    'grid.alpha': 0.5,
    'xtick.color': 'black',
    'ytick.color': 'black',
    'figure.facecolor': 'white',
    'figure.edgecolor': 'white'
})

# sns.set_theme('seaborn-whitegrid')


# Ścieżki do katalogów z obrazami i anotacjami
# image_dir = 'datasets/daneMieszane/train/images/'
# label_dir = 'datasets/daneMieszane/train/labels'
path= "datasets/daneZRobota/"

# Lista plików anotacji
annotation_files = [f for f in os.listdir(path+"/train/labels") if f.endswith('.txt')]
annotation_files2 = [f for f in os.listdir(path+"val/labels") if f.endswith('.txt')]


# annotation_files.append(annotation_files2)
# Zmienne do przechowywania danych
annotations = []
image_sizes = {}

klasy = {0: "Człowiek", 1:"Hulajnoga", 2:"Samochód"}

# Przetwarzanie plików anotacji
for annotation_file in annotation_files:
    image_id = os.path.splitext(annotation_file)[0]
    # image_format = os.path.splitext(annotation_file)[1]
    
    # Ścieżka do pliku anotacji
    annotation_path = os.path.join(path+"/train/labels", annotation_file)
    # Wczytywanie anotacji
    with open(annotation_path, 'r') as f:
        for line in f.readlines():
            category_id, x_center, y_center, w, h = map(float, line.strip().split())
            annotations.append((image_id, klasy[int(category_id)], x_center, y_center, w, h))


# Przetwarzanie plików anotacji
for annotation_file in annotation_files2:
    image_id = os.path.splitext(annotation_file)[0]
    # image_format = os.path.splitext(annotation_file)[1]
    
    # Ścieżka do pliku anotacji
    annotation_path = os.path.join(path+"/val/labels", annotation_file)
    # Wczytywanie anotacji
    with open(annotation_path, 'r') as f:
        for line in f.readlines():
            category_id, x_center, y_center, w, h = map(float, line.strip().split())
            annotations.append((image_id, klasy[int(category_id)], x_center, y_center, w, h))



# Konwersja do DataFrame
df_annotations = pd.DataFrame(annotations, columns=['image_id', 'category_id', 'x_center', 'y_center', 'width', 'height'])
# df_image_sizes = pd.DataFrame(image_sizes.items(), columns=['image_id', 'size'])
# df_image_sizes[['width', 'height']] = pd.DataFrame(df_image_sizes['size'].tolist(), index=df_image_sizes.index)

# 1. Liczba obrazów
# num_images = df_image_sizes['image_id'].nunique()
# print(f'Liczba obrazów: {num_images}')

# 2. Liczba kategorii
num_categories = df_annotations['category_id'].nunique()
print(f'Liczba kategorii: {num_categories}')

# 3. Liczba anotacji
num_annotations = len(df_annotations)
print(f'Liczba anotacji: {num_annotations}')

# 4. Anotacje na kategorię
category_annotation_counts = df_annotations['category_id'].value_counts().reset_index()
category_annotation_counts.columns = ['Kategoria', 'Liczba annotacji']

# 5. Wizualizacja liczby anotacji na kategorię
plt.figure(figsize=(12, 8))

# sns.barplot(x='Category', y='Liczba annotacji', data=category_annotation_counts)
palette = sns.color_palette("viridis", num_categories)
sns.barplot(x='Kategoria', y='Liczba annotacji', data=category_annotation_counts, palette=palette)

# colors = {'Czlowiek':'red', 'Hulajnoga':'green', "Samochód": "yellow"}
# labels = list(colors.keys())
# handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
# plt.legend(handles, labels)

plt.xticks(rotation=90)
plt.title('Liczba anotacji na kategorię')
plt.show()
```
